### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
```

```python
sales = pd.read_csv("./data/company_sales_data.csv")
sales
```

```python
x = sales["month_number"]
totals = sales["total_profit"]

plt.plot(x, totals)
plt.title("Total Profit by Month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
```

### Exercise 2. 

Get total units of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

```python
units = sales["total_units"]

plt.plot(x, units, 'o', linestyle=":", color='red', linewidth=3, label='# Units')
plt.title("Number of Units Sold Per Month")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc='lower right')
```

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
fig = plt.figure(figsize=(11, 5))
plt.plot(x, sales.facecream, label="Facecream")
plt.plot(x, sales.facewash, label="Facewash")
plt.plot(x, sales.toothpaste, label="Toothpaste")
plt.plot(x, sales.bathingsoap, label="Bathing Soap")
plt.plot(x, sales.shampoo, label="Shampoo")
plt.plot(x, sales.moisturizer, label="Moisturizer")
plt.legend(loc="upper left")
plt.title("Amount of Each Product Sold Per Month")
plt.xlabel("Month Number")
plt.ylabel("Number Units Sold")
```
